```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

import plotly as py
import cufflinks as cf
from plotly.offline import iplot
import plotly.express as px
```

```python
py.offline.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
# 3d plot the data using matplotlib
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection = '3d')
ax.scatter3D(xs = data['sepal_length'], ys = data['sepal_width'], zs = data['petal_length'], c = data['petal_width'])
plt.show()
```

```python
px.scatter_3d(x = data['sepal_length'], y = data['sepal_width'], z = data['petal_length'], color = data['petal_width'], size_max = 10,)
```

```python
px.line_3d(x = data['sepal_length'], y = data['sepal_width'], z = data['petal_length'], color = data['petal_width'], width = 600, height = 600)
```

```python
data1 = data.drop(['species'], axis = 1)
data1.iplot(kind = 'surface')
```
