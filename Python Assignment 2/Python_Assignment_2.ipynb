```python
# Solution1
def printpattern(num):
    for i in range(1, num +1):
        for j in range(1, i +1):
            print("*",end = "")
        print("\r")

    for i in range(1, num):
        for k in reversed(range(num-i)):
            print("*", end="")
        print("\r")




if __name__ == "__main__":
  inputnumber = int(input("Enter the maximum star in inverted pyramid : "))
  printpattern(inputnumber)
```
