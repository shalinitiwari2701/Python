```python
class Triangle_Parameters:
  def __init__(self, a,b,c):
    self.a = a
    self.b = b
    self.c = c

class Calculate_Areaof_Triangle(Triangle_Parameters):
  def __init__(self, *args):
    super(Calculate_Areaof_Triangle, self).__init__(*args)

  def calculatearea(self):
    s = (self.a +self.b +self.c)/2
    area = (s * (s-self.a) * (s-self.b) * (s- self.c)) ** .5
    return area


a = int(input("Enter side of triangle a : " ))
b = int(input("Enter side of triangle b : " ))
c = int(input("Enter side of triangle c : " ))
area = Calculate_Areaof_Triangle(a,b,c)
print("Area of triangle is :" , area.calculatearea())
```

```python
#1.2 Solution
def filter_long_words(inputlist, n):
  result = []
  for i in inputlist:
    if len(i) > n:
      result.append(i)
  return result

filter_long_words(["sdfd","tuuiodsjfdjfdjfd","shalini"], 5)
```

```python
#2.1 Solution

def listofwords_to_lentowords(inputlist):
  ''' This is a function that maps list of words into a list of integers
representing the lengths of the corresponding words.'''
  result = []
  for i in inputlist:
    result.append(len(i))
  return result

listofwords_to_lentowords(["sdfd","tuuiodsjfdjfdjfd","shalini"])
```

```python
#2.2 Solution

def Verify_IfCharacter_IsVowel(s):
  vowellist = ['a', 'e', 'i', 'o', 'u']
  for i in vowellist:
    if s == i:
      return True
    else:
      return False

character = input("enter the character to find whether it is vowel or not : ")
Verify_IfCharacter_IsVowel(character)
```
