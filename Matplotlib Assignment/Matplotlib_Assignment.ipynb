```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from plotly.offline import iplot
import cufflinks as cf
import plotly.tools as tls
import plotly as py
```

```python
py.offline.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
lst1 = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
```

```python
lst2 = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
max = pd.Series(lst1, name = 'Max')
min = pd.Series(lst2, name = 'Min')
```

```python
temp_data = pd.concat([max, min], axis=1)
temp_data.head()
```

```python
temp_data.plot(xlabel = 'Month', ylabel = 'Temperature', kind = 'line')
```

```python
sns.relplot(data = temp_data, kind = 'line', marker = "o", dashes = True)
```

```python
sns.lineplot(data = temp_data, marker = "o", dashes = True)
```

```python
temp_data.iplot( mode = 'markers' )
```
