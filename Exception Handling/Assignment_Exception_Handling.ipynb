```python
# solution1

def Checkif_Division_IsPossible():
  try :
    a = int(input("Enter the integer value "))
    b = int(input("Enter another integer number "))
    c = a/b
    print("Division result is ", c)
  except ZeroDivisionError as e:
    print("this is zero division error")
    print(e)

  except ValueError as e:
    print("This is input output error that may occur when input is not int")
    print(e)

```

```python
Checkif_Division_IsPossible()
```
