**# Solution 1 Myreduce function**

```python
def add(x, y):
  return x + y


# output
def myreturn(functionname, listname):
  output = listname[0]     # previously I have assigned this value to 0, it was giving wrong result with multipliation and substraction
  for i in listname[1:]:
    output = functionname(output, i)

  return output
```

```python
print(myreturn(add, [1,2,3,4]))
```

```python
print(myreturn(lambda x,y : x + y, [1,2,3,4]))
```

```python
print(myreturn(lambda x,y : x*y, [1,2,3,4]))
```

**#Solution2 myfilter function**

```python
#Solution2
l = [12,18,45.67,78,99,True, 2+5j, False,[1,2,3,4],5.5,"Shalini",2444,[5,6,7,8,9],"Ineuron"]
```

```python
def IdentifyInttype(number):
   updatedlist = []
   for i in number:
     if type(i) == int:
       updatedlist.append(i)
   return updatedlist
```

```python
def myfilter(FunctionName, Iterablelist):
  return FunctionName(Iterablelist)
```

```python
result11 = myfilter(IdentifyInttype,l)
```

```python
print(result11)
```

```python
#Solution3
#['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']
```
